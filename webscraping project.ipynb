```
from bs4 import BeautifulSoup
import requests
```

```
url = 'https://www.cricbuzz.com/cricket-series/7476/icc-mens-t20-world-cup-2024/stats'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```
print(soup)
```

```
soup.find_all("table")
```

```
soup.find_all("table")[0]
```

```
soup.find('table', class_ = 'table table-responsive cb-series-stats')
```

```
table = soup.find_all('table')[0]
```

```
print(table)
```

```
players_titles=soup.find_all("th")
```

```
players_titles
```

```
players_title_table=[title.text for title in players_titles]
print(players_title_table)
```

```
players_title_table.pop(0)
```

```
players_title_table
```

```
import pandas as pd
df=pd.DataFrame(columns=players_title_table)
print(df)
```

```
column=table.find_all('tr')
```

```
for row in column[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)

    length = len(df)
    df.loc[length] = individual_row_data
```
